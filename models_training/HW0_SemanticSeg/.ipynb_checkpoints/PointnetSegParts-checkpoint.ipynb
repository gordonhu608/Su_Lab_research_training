```python
import torch
from torch import nn
import torch.nn.functional as F
import numpy as np
from torch.utils.data import DataLoader
import os
```

```python
file_ids = [f.split('.')[0] for f in os.listdir("train/pts")]
```

```python
class TrainDataset(torch.utils.data.Dataset):
    def __init__(self):
        super().__init__()
        pts = []
        labels = []
        for id in file_ids:
            with open(f'train/pts/{id}.pts') as f:
                p = []
                for line in f:
                    p.append(np.array([float(x) for x in line.split()]))
                pts.append(np.array(p))
            with open(f'train/label/{id}.txt') as f:
                p = []
                for line in f:
                    p.append(float(line))
                labels.append(np.array(p))
        maxlen = max([len(x) for x in pts])
        pad_pts = np.zeros((len(pts), maxlen, 3))
        pad_labels = np.zeros((len(pts), maxlen))
        for i, (p, l) in enumerate(zip(pts, labels)):
            pad_pts[i, :len(p)] = p
            pad_pts[i, len(p):] = p[-1]
            pad_labels[i, :len(l)] = l
            pad_labels[i, len(l):] = l[-1]
            
        self.pts = pad_pts
        self.labels = (pad_labels - 1).astype(np.long)
        
    def __len__(self):
        return len(self.pts)
    def __getitem__(self, idx):
        pts = self.pts[idx]
        labels = self.labels[idx]
        
        pts = torch.tensor(pts.T).float().contiguous()
        labels = torch.tensor(labels).long().contiguous()
        return {"pts": pts, "labels": labels}
```

```python
class TestDataset(torch.utils.data.Dataset):
    def __init__(self):
        pts = []
        for id in range(6):
            with open(f'test/{id}.pts') as f:
                p = []
                for line in f:
                    p.append(np.array([float(x) for x in line.split()]))
                pts.append(np.array(p))
        maxlen = max([len(x) for x in pts])
        pad_pts = np.zeros((len(pts), maxlen, 3))
        for i, p in enumerate(pts):
            pad_pts[i, :len(p)] = p
            pad_pts[i, len(p):] = p[-1]
        super().__init__()
        self.pts = pad_pts
    def __len__(self):
        return len(self.pts)
    def __getitem__(self, idx):
        pts = self.pts[idx]
        pts = torch.tensor(pts.T).float().contiguous()
        return pts
```

```python
class PointNet(nn.Module):
    def __init__(self):
        super(PointNet, self).__init__()
        self.conv1 = torch.nn.Conv1d(3, 64, 1)
        self.conv2 = torch.nn.Conv1d(64, 128, 1)
        self.conv3 = torch.nn.Conv1d(128, 256, 1)
        self.bn1 = nn.BatchNorm1d(64)
        self.bn2 = nn.BatchNorm1d(128)
        self.bn3 = nn.BatchNorm1d(256)
        self.dense1 = torch.nn.Linear(256, 256)
        self.bn4 = nn.BatchNorm1d(256)
        self.dense2 = torch.nn.Linear(256, 256)
        self.bn5 = nn.BatchNorm1d(256)
        self.dense3 = torch.nn.Linear(256, 256)
        self.fconv1 = torch.nn.Conv1d(512, 128, 1)
        self.fconv2 = torch.nn.Conv1d(128, 64, 1)
        self.fconv3 = torch.nn.Conv1d(64, 4, 1)
        self.fbn1 = nn.BatchNorm1d(128)
        self.fbn2 = nn.BatchNorm1d(64)
        for net in [
            self.conv1,
            self.conv2,
            self.conv3,
            self.dense1,
            self.dense2,
            self.dense3,
            self.fconv1,
            self.fconv2,
            self.fconv3,
        ]:
            torch.nn.init.xavier_uniform_(net.weight)
    def forward(self, x):
        points = x 
        mins = points.min(dim=2, keepdim=True).values
        maxs = points.max(dim=2, keepdim=True).values
        center = (mins + maxs) / 2 # N, 3, 1
        half_extents = (maxs - mins) / 2 # N, 3, 1
        longest = half_extents.max(dim=1, keepdim=True).values.clamp(
        min=1e-3
        ) # N, 1, 1
        points = (points - center) / longest
        x = points
        x = F.relu(self.bn1(self.conv1(x)))
        x = F.relu(self.bn2(self.conv2(x)))
        x = self.bn3(self.conv3(x))
        y = x # N, 256, 2907 
        x = torch.max(x, 2, keepdim=True)[0] # 64, 256, 1
        x = x.view(-1, 256) # 64, 256 
        x = F.relu(self.bn4(self.dense1(x)))
        x = F.relu(self.bn5(self.dense2(x)))
        x = self.dense3(x)   # 64, 256 
        z = x[..., None].expand((-1, x.shape[1], y.shape[-1])) # 64, 256, 2907
        x = torch.cat([y, z], dim=1) 
        x = F.relu(self.fbn1(self.fconv1(x)))
        x = F.relu(self.fbn2(self.fconv2(x)))
        x = self.fconv3(x)
        return x
```

```python
import open3d
dataset = TrainDataset()
train_loader = DataLoader(dataset, batch_size=64, shuffle=True)
model = PointNet()
model.cuda()
```

```python
optim = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
from tqdm.notebook import tqdm
total_count = 0
total_loss = 0

for epoch in tqdm(range(1000)):
    for param_group in optim.param_groups:
        param_group['lr'] = param_group['lr'] * 0.99
    for step, d in enumerate(train_loader):
        points = d['pts'].cuda()
        labels = d['labels'].cuda()
        pred = model(points)
        loss = F.cross_entropy(pred, labels)
        
        total_loss += loss.item()
        total_count += 1
        
        optim.zero_grad()
        loss.backward()
        optim.step()
    if epoch % 10 == 0:
        print(f'epoch {epoch+1} step {step}; loss: {total_loss / total_count}')
    total_count = 0
    total_loss =0
```

```python
model.eval()
test_dataset = TestDataset()
```

```python
preds = []
for x in test_dataset: 
    x = x[None].cuda()
    y = model(x)
    preds.append(y.argmax(1)[0].cpu().numpy())
```

```python
import torch.utils.data
from torch.autograd import Variable
class STN3d(nn.Module):
    def __init__(self):
        super(STN3d, self).__init__()
        self.conv1 = torch.nn.Conv1d(3, 64, 1)
        self.conv2 = torch.nn.Conv1d(64, 128, 1)
        self.conv3 = torch.nn.Conv1d(128, 1024, 1)
        self.fc1 = nn.Linear(1024, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 9)
        self.relu = nn.ReLU()

        self.bn1 = nn.BatchNorm1d(64)
        self.bn2 = nn.BatchNorm1d(128)
        self.bn3 = nn.BatchNorm1d(1024)
        self.bn4 = nn.BatchNorm1d(512)
        self.bn5 = nn.BatchNorm1d(256)


    def forward(self, x):
        batchsize = x.size()[0]
        x = F.relu(self.bn1(self.conv1(x)))
        x = F.relu(self.bn2(self.conv2(x)))
        x = F.relu(self.bn3(self.conv3(x)))
        x = torch.max(x, 2, keepdim=True)[0]
        x = x.view(-1, 1024)

        x = F.relu(self.bn4(self.fc1(x)))
        x = F.relu(self.bn5(self.fc2(x)))
        x = self.fc3(x)

        iden = Variable(torch.from_numpy(np.array([1,0,0,0,1,0,0,0,1]).astype(np.float32))).view(1,9).repeat(batchsize,1)
        iden = iden.cuda()
        x = x + iden
        x = x.view(-1, 3, 3)
        return x
```

```python
class PointNetfeat(nn.Module):
    def __init__(self, global_feat = True, feature_transform = False):
        super(PointNetfeat, self).__init__()
        self.stn = STN3d()
        self.conv1 = torch.nn.Conv1d(3, 64, 1)
        self.conv2 = torch.nn.Conv1d(64, 128, 1)
        self.conv3 = torch.nn.Conv1d(128, 1024, 1)
        self.bn1 = nn.BatchNorm1d(64)
        self.bn2 = nn.BatchNorm1d(128)
        self.bn3 = nn.BatchNorm1d(1024)
        self.global_feat = global_feat
        self.feature_transform = feature_transform
        if self.feature_transform:
            self.fstn = STNkd(k=64)

    def forward(self, x):
        n_pts = x.size()[2]
        trans = self.stn(x)
        x = x.transpose(2, 1)
        x = torch.bmm(x, trans)
        x = x.transpose(2, 1)
        x = F.relu(self.bn1(self.conv1(x)))

        if self.feature_transform:
            trans_feat = self.fstn(x)
            x = x.transpose(2,1)
            x = torch.bmm(x, trans_feat)
            x = x.transpose(2,1)
        else:
            trans_feat = None

        pointfeat = x
        x = F.relu(self.bn2(self.conv2(x)))
        x = self.bn3(self.conv3(x))
        x = torch.max(x, 2, keepdim=True)[0]
        x = x.view(-1, 1024)
        if self.global_feat:
            return x, trans, trans_feat
        else:
            x = x.view(-1, 1024, 1).repeat(1, 1, n_pts)
            return torch.cat([x, pointfeat], 1), trans, trans_feat

class PointNetSeg(nn.Module):
    def __init__(self, k = 2, feature_transform=False):
        super(PointNetSeg, self).__init__()
        self.k = k
        self.feature_transform=feature_transform
        self.feat = PointNetfeat(global_feat=False, feature_transform=feature_transform)
        self.conv1 = torch.nn.Conv1d(1088, 512, 1)
        self.conv2 = torch.nn.Conv1d(512, 256, 1)
        self.conv3 = torch.nn.Conv1d(256, 128, 1)
        self.conv4 = torch.nn.Conv1d(128, self.k, 1)
        self.bn1 = nn.BatchNorm1d(512)
        self.bn2 = nn.BatchNorm1d(256)
        self.bn3 = nn.BatchNorm1d(128)

    def forward(self, x):
        batchsize = x.size()[0]
        n_pts = x.size()[2]
        x, trans, trans_feat = self.feat(x)
        x = F.relu(self.bn1(self.conv1(x)))
        x = F.relu(self.bn2(self.conv2(x)))
        x = F.relu(self.bn3(self.conv3(x)))
        x = self.conv4(x)
        x = x.transpose(2,1).contiguous()
        x = F.log_softmax(x.view(-1,self.k), dim=-1)
        x = x.view(batchsize, n_pts, self.k)
        return x, trans, trans_feat

def feature_transform_regularizer(trans):
    d = trans.size()[1]
    batchsize = trans.size()[0]
    I = torch.eye(d)[None, :, :]
    I = I.cuda()
    loss = torch.mean(torch.norm(torch.bmm(trans, trans.transpose(2,1)) - I, dim=(1,2)))
    return loss
```

```python
model = PointNetSeg(4, False)
model.cuda()
```

```python
total_count = 0
total_loss = 0

for epoch in tqdm(range(100)):
    for param_group in optim.param_groups:
        param_group['lr'] = param_group['lr'] * 0.99
    for step, d in enumerate(train_loader):
        points = d['pts'].cuda()
        labels = d['labels'].cuda()
        pred = model(points)
        loss = F.cross_entropy(pred, labels)
        
        total_loss += loss.item()
        total_count += 1
        
        optim.zero_grad()
        loss.backward()
        optim.step()
    if epoch % 10 == 0:
        print(f'epoch {epoch+1} step {step}; loss: {total_loss / total_count}')
    total_count = 0
    total_loss =0
```

```python
model.eval()
test_dataset = TestDataset()
```

```python
preds = []
for x in test_dataset: 
    x = x[None].cuda()
    y = model(x)
    preds.append(y.argmax(1)[0].cpu().numpy())
```
